```
%matplotlib inline
from sklearn.datasets import load_iris
from mpl_toolkits.mplot3d import Axes3D
from sklearn.decomposition import PCA
from pandas.plotting import radviz

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```
dataset = load_iris()
df = pd.DataFrame(dataset.data, columns=dataset.feature_names)
df['labels'] = dataset.target
df.head()
```

```
print(df.size)
print(df.info())
df.describe()
```

```
df.groupby('labels').mean()
```

```
df['sepal length (cm)'].hist(bins=30)
```

```
for class_number in np.unique(dataset.target):
    plt.figure(1)
    df['sepal length (cm)'].iloc[np.where(dataset.target == class_number)[0]].hist(bins=30)
```

```
fig, ax = plt.subplots()
df.plot(kind='scatter', x='sepal length (cm)', y='sepal width (cm)', s=60, c='labels', cmap='viridis', ax=ax) 
```

```
radviz(df, 'labels')
```

scikit-learn example
ref: https://scikit-learn.org/stable/auto_examples/datasets/plot_iris_dataset.html

```
X = dataset.data[:, :2]  # we only take the first two features.
y = dataset.target
```

```
x_min, x_max = X[:, 0].min() - .5, X[:, 0].max() + .5
y_min, y_max = X[:, 1].min() - .5, X[:, 1].max() + .5

plt.figure(2, figsize=(8, 6))
plt.clf()

# Plot the training points
plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.Set1,
            edgecolor='k')
plt.xlabel('Sepal length')
plt.ylabel('Sepal width')

plt.xlim(x_min, x_max)
plt.ylim(y_min, y_max)
plt.xticks(())
plt.yticks(())
```

```
# To getter a better understanding of interaction of the dimensions
# plot the first three PCA dimensions
fig = plt.figure(1, figsize=(8, 6))
ax = Axes3D(fig, elev=-150, azim=110)
X_reduced = PCA(n_components=3).fit_transform(dataset.data)
ax.scatter(X_reduced[:, 0], X_reduced[:, 1], X_reduced[:, 2], c=y,
           cmap=plt.cm.Set1, edgecolor='k', s=40)
ax.set_title("First three PCA directions")
ax.set_xlabel("1st eigenvector")
ax.w_xaxis.set_ticklabels([])
ax.set_ylabel("2nd eigenvector")
ax.w_yaxis.set_ticklabels([])
ax.set_zlabel("3rd eigenvector")
ax.w_zaxis.set_ticklabels([])

plt.show()
```
